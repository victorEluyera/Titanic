```python
df[df['parch'].isna()]
df['parch'].fillna('Unknown value', inplace=True)
df['parch'].value_counts()
```

```python
def rearrange_name(full_name):
    parts = full_name.split(', ')
    if len(parts) == 2 and '.' in parts[1]:
        title = parts[1].split('.')[0] + '.'
        first_name = parts[1].split('.')[1].strip()
        last_name = parts[0]
        return f"{title} {first_name} {last_name}"
    else:
        return full_name  

# Apply the function to the 'name' column
df['name'] = df['name'].apply(rearrange_name)
df
```

## How many People Survived?


```python
survival=df['survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survival, labels=['Not Survived', 'Survived'], autopct='%1.1f%%')
plt.title('Titanic Survival')
plt.show()
# plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
```
